**data source: https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page**

Sample data from Parquet files in Pandas dataframe. Then, create a CSV file to use for data cleaning.

```python
# create paths for each month's parquet file
jan = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-01.parquet"
feb = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-02.parquet"
mar = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-03.parquet"
apr = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-04.parquet"
may = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-05.parquet"
june = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-06.parquet"
july = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-07.parquet"
aug = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-08.parquet"
sept = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-09.parquet"
oct = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-10.parquet"
nov = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-11.parquet"
dec = "C:/Files_JRR/Projects/project_2_data/trip_data_by_month/fhvhv_tripdata_2023-12.parquet"

# create an individual df for each parquet file
jan_df = pd.read_parquet(jan)
feb_df = pd.read_parquet(feb)
mar_df = pd.read_parquet(mar)
apr_df = pd.read_parquet(apr)
may_df = pd.read_parquet(may)
june_df = pd.read_parquet(june)
july_df = pd.read_parquet(july)
aug_df = pd.read_parquet(aug)
sept_df = pd.read_parquet(sept)
oct_df = pd.read_parquet(oct)
nov_df = pd.read_parquet(nov)
dec_df = pd.read_parquet(dec)

# create a sample df of 100 trips per month
jan_df_sampled = jan_df.sample(100)
feb_df_sampled = feb_df.sample(100)
mar_df_sampled = mar_df.sample(100)
apr_df_sampled = apr_df.sample(100)
may_df_sampled = may_df.sample(100)
june_df_sampled = june_df.sample(100)
july_df_sampled = july_df.sample(100)
aug_df_sampled = aug_df.sample(100)
sept_df_sampled = sept_df.sample(100)
oct_df_sampled = oct_df.sample(100)
nov_df_sampled = nov_df.sample(100)
dec_df_sampled = dec_df.sample(100)

# concat df
df = pd.concat([jan_df_sampled, feb_df_sampled, mar_df_sampled, apr_df_sampled, may_df_sampled, june_df_sampled, july_df_sampled, aug_df_sampled, sept_df_sampled, oct_df_sampled, nov_df_sampled, dec_df_sampled])

# set the pickup time as the index
df.set_index("request_datetime", inplace=True)

# sort by the index
df.sort_index(inplace=True)

# save as csv
df.to_csv("trip_data.csv")

# preview df
df
```
